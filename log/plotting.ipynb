```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#fname_cpu = 'eth_mnist_cpu.npz'
fname_cpu = 'eth_mnist_cpu_000.npz'
data_cpu = np.load(fname_cpu)
data_cpu = data_cpu['data']
all_latest_cpu = data_cpu[:,0]
all_avg_cpu = data_cpu[:,1]
all_best_cpu = data_cpu[:,2]
prop_latest_cpu = data_cpu[:,3]
prop_avg_cpu = data_cpu[:,4]
prop_best_cpu = data_cpu[:,5]
time_usage_cpu = data_cpu[:,6]
```

```python
figure = plt.figure(figsize=[16,6])
#plt.subplot(3,1,1)
plt.plot(all_latest_cpu,'-',label='all_latest_cpu')
plt.plot(all_avg_cpu,'-',label='all_avg_cpu')
plt.plot(all_best_cpu,'-',label='all_best_cpu')

#plt.subplot(3,1,2)
plt.plot(prop_latest_cpu,'-',label='prop_latest_cpu')
plt.plot(prop_avg_cpu,'-',label='prop_avg_cpu')
plt.plot(prop_best_cpu,'-',label='prop_best_cpu')
plt.legend()
```

```python
#fname_gpu = 'eth_mnist_gpu.npz'
fname_gpu = 'eth_mnist_gpu_005.npz'
data_gpu = np.load(fname_gpu)
data_gpu = data_gpu['data']
all_latest_gpu = data_gpu[:,0]
all_avg_gpu = data_gpu[:,1]
all_best_gpu = data_gpu[:,2]
prop_latest_gpu = data_gpu[:,3]
prop_avg_gpu = data_gpu[:,4]
prop_best_gpu = data_gpu[:,5]
time_usage_gpu = data_gpu[:,6]
```

```python
figure = plt.figure(figsize=[16,6])
plt.subplot(3,1,(1,2))
plt.plot(all_latest_gpu,'-',label='all_latest_gpu')
plt.plot(all_avg_gpu,'-',label='all_avg_gpu')
plt.plot(all_best_gpu,'-',label='all_best_gpu')

plt.plot(prop_latest_gpu,'-',label='prop_latest_gpu')
plt.plot(prop_avg_gpu,'-',label='prop_avg_gpu')
plt.plot(prop_best_gpu,'-',label='prop_best_gpu')
plt.legend()
plt.subplot(3,1,3)
plt.plot(time_usage_gpu,'-',label='average-time-usage-each-sample')
plt.legend()
```

```python
figure = plt.figure(figsize=[16,6])
plt.subplot(3,1,(1,2))
plt.plot(all_latest_gpu,'-',label='all_latest_gpu')
plt.plot(all_avg_gpu,'-',label='all_avg_gpu')
plt.plot(all_best_gpu,'-',label='all_best_gpu')


plt.plot(all_latest_cpu,'-',label='all_latest_cpu')
plt.plot(all_avg_cpu,'-',label='all_avg_cpu')
plt.plot(all_best_cpu,'-',label='all_best_cpu')
plt.legend()
plt.subplot(3,1,3)
plt.plot(time_usage_cpu,'-',label='cpu-time usage')
plt.plot(time_usage_gpu,'-',label='gpu-time usage')
plt.legend()
```
