```python
from bindsnet.encoding import poisson
import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
train_time = 200
rest_time = 100
rate_base = 10.
n_neuron = 100
p = poisson(torch.tensor(range(n_neuron)).float()*rate_base, train_time)
plt.matshow(p)
plt.show()
```

```python
p
```

```python
rest_p = torch.tensor(np.zeros((rest_time,n_neuron),dtype=np.uint8))
```

```python
rest_p.shape
```

```python
final_p = torch.cat((p,rest_p),0)
```
